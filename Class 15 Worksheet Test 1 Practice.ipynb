### Problem 3
A group of friends have decided to start a club. The club name will be the first letter of each of their names in the order in which they're given.  The club name should be all uppercase, regardless of how the name is given.

Write code that, given a list of names called *names*, creates a string called *club_name* that is the name of the club.  Your code should work for any length list of names.

````python
names = ["Bertha", "agatha", "Thomas"]

````

````python
#solution
names = ["Bertha", "agatha", "Thomas"]

club_name = ""
for name in names:
    club_name += name[0].upper()
    #club_name = club_name + name[0].upper()
    
club_name    
````

### Problem 4
You are given an integer *myint*.  Add code to the cell below to compute the sum and product of the digits in *myint*.  Call the new quantities *mysum* and *myproduct*.

````python
myint = 457
````

````python
#solution
myint = 457

mysum = 0
myproduct = 1

for digit in str(myint): 
    mysum += int(digit)
    myproduct *= int(digit) # or product = product * int(digit)
    
mysum
myproduct
````
